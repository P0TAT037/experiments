# Text Detection

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('./sample.jpg')

f, ((a, b, c), (d, e, f)) = plt.subplots(2, 3, figsize=(20, 20))

a.set_title('Original')
a.imshow(img)

# Convert to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
grayscale = img.copy()
b.set_title('Grayscale')
b.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# Apply Canny's edge detection
img = cv2.Canny(img, 40, 80)
c.set_title('Canny')
c.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# Apply dilation
kernel = np.ones((3, 3), np.uint8)
img = cv2.dilate(img, kernel)
d.set_title('Dilation')
d.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# Fill text area with white pixels

# cv2.floodFill(img, None, (0, 0), 255)
contours, hierarchy = cv2.findContours(image=img, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img, [max(contours, key = cv2.contourArea)], -1, 255, thickness=-1) # @loay212 افهم وفهمني هه
e.set_title('Floodfill')
e.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# select the biggest white area

def detectTextArea(img): #crop the biggest white area in the picture (@loay212 implement this plz <3)
    return img

img = detectTextArea(grayscale)
f.set_title('Text area')
f.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
```
